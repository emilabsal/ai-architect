```python
import matplotlib.pyplot as plt
```

```python
data = {
    'Survived': [1, 0, 1, 0, 1],
    'Pclass': [1, 2, 3, 1, 3],
    'Name': ['John Smith', 'Jane Doe', 'Michael Johnson', 'Emily Brown', 'David Wilson'],
    'Sex': ['male', 'female', 'male', 'female', 'male'],
    'Age': [30, 25, 40, 35, 20],
    'SibSp': [1, 0, 1, 0, 0],
    'Parch': [0, 1, 2, 0, 0],
    'Ticket': ['A1234', 'B5678', 'C91011', 'D1213', 'E1415'],
    'Fare': [100.50, 50.25, 75.75, 200.00, 30.50],
    'Cabin': ['A1', 'B2', 'C3', 'D4', 'E5'],
    'Embarked': ['S', 'C', 'S', 'S', 'C']
}
#Интервалы
age_intervals = [0, 15, 30, 45, 60, 75]
#списки для хранения количества выживших и умерших в каждом возрастном интервале
survived_counts = [0] * len(age_intervals)
dead_counts = [0] * len(age_intervals)
#Считаем количество выживших и умерших в каждом возрастном интервале
for i in range(len(data['Survived'])):
    age = data['Age'][i]
    for j in range(len(age_intervals)):
        if age <= age_intervals[j]:
            if data['Survived'][i] == 1:
                survived_counts[j] += 1
            else:
                dead_counts[j] += 1
            break
#визуализация
plt.bar(range(len(age_intervals)), survived_counts, label='Survived', color='green')
plt.bar(range(len(age_intervals)), dead_counts, bottom=survived_counts, label='Dead', color='red')
plt.xticks(range(len(age_intervals)), age_intervals)
plt.xlabel('Age Intervals')
plt.ylabel('Count')
plt.title('Survival by Age')
plt.legend()
plt.show()
```
